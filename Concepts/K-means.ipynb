```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
print('Libraries ok!')
```

```python
dataset= np.array([[1,4],[2,2],[2,5],[3,3],[3,5],[4,7],[5,6],[6,4],[6,7],[7,6],[7,9],[8,7],[8,9],[9,4],[9,8]])
```

```python
plt.scatter(dataset[:,0],dataset[:,1] )
plt.xlim(0,10)
plt.ylim(0,10)

plt.grid()
```

```python
# Utilizando kmenas encontramos 3 centroides... initi define a estratégia... o "k-means++" é um algoritmo interno
# mais assertivo... também usamos o "n_init"
```

```python
kmeans = KMeans(n_clusters = 3, init='k-means++', max_iter=300, n_init=10)
pred_y = kmeans.fit_predict(dataset)

# edu -> esse pred_y de alguma maneira gera o "kmeans.cluster_centers"... se comentar a linha dá erro!!!
```

```python
plt.scatter(dataset[:,0],dataset[:,1] )
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s=70,c='red')
plt.show()
```

```python
# exemplo com dados aleatórios
```

```python
X,y = make_blobs(n_samples=500, centers=50,random_state=0)
```

```python
# edu... perceba que usa apenas o X

plt.scatter(X[:,0], X[:,1])
```

```python
wcss = []
for i in range(1,20):
    kmeans = KMeans(n_clusters = i, init='k-means++', max_iter=300, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # soma dos quadrados...
```

```python
plt.plot(range(1,20), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid() # edu...
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=pred_y)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s=50,c='red')
plt.show()
```
